```python
import pandas as pd
```

```python
customers_df = pd.read_csv("customer_data_.csv")
transactions_df = pd.read_csv("transaction_data_.csv")
```

```python
print(customers_df.shape)
print(transactions_df.shape)
```

```python
customers_df["Customer ID"] = customers_df["Customer ID"].astype(int)
transactions_df["Customer ID"] = transactions_df["Customer ID"].astype(int)
```

```python
merged_df = pd.merge(
    transactions_df,
    customers_df,
    on="Customer ID",
    how="inner"
)
```

```python
print("merged_df shape:", merged_df.shape)
merged_df.head()
```

```python
gender_summary = merged_df.groupby("Gender").agg({
    "Quantity": "sum",
    "Total Price": "sum"
})
```

```python
print("gender_summary shape:", gender_summary.shape)
gender_summary
```

```python
gender_summary["Total Price"].plot(
    kind="bar",
    figsize=(8,4),
    title="Total Sales by Gender",
    color="lightgreen"
)
```

```python
country_summary = merged_df.groupby('Country').agg({
    'Quantity': 'sum',
    'Total Price': 'sum'
})

country_summary
```

```python
country_summary['Total Price'].plot(
    kind='bar',
    figsize=(8,4),
    title='Total Sales by Country',
    color='lightblue'
)
```

```python
gender_summary = merged_df.groupby('Gender').agg({
    'Quantity': 'sum',
    'Total Price': 'sum'
})

gender_summary
```

```python
gender_summary['Quantity'].plot(
    kind='bar',
    figsize=(8,4),
    title='Quantity Sold by Gender',
    color='blue'
)
```

```python
high_value_transactions = merged_df[merged_df['Total Price'] > 1000]
high_value_transactions
```

```python
high_value_transactions.shape
```

```python
product_summary = merged_df.groupby('Product').agg({
    'Quantity': 'sum',
    'Total Price': 'sum'
})

product_summary
```

```python
product_summary['Total Price'].plot(
    kind='bar',
    figsize=(8,4),
    title='Total Sales by Product',
    color='purple'
)
```

```python
product_summary['Quantity'].plot(
    kind='bar',
    figsize=(8,4),
    title='Quantity Sold by Product',
    color='brown'
)
```
